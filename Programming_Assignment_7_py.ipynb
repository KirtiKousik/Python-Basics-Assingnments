2. Write a Python Program to find largest element in an array?

```python
def largest(arr, n):
    max = arr[0]
    for i in range(1, n):
        if arr[i] > max:
            max = arr[i]
    return max

arr = [12,4,6,13]
n = len(arr)
ans = largest(arr, n)
print(f"The Largest number in array {arr} is {ans}")
```

3. Write a Python Program for array rotation?

```python
def rotateArray(arr, n, d):
    temp = []
    i = 0
    while (i < d):
        temp.append(arr[i])
        i = i + 1
    i = 0
    while (d < n):
        arr[i] = arr[d]
        i = i + 1
        d = d + 1
    arr[:] = arr[: i] + temp
    return arr

arr = [1, 2, 3, 4, 5, 6, 7]
print('The array before ratation is: \n',arr)
print('Array after left rotation by 2 is:')
print(rotateArray(arr, len(arr), 2))
```

4. Write a Python Program to Split the array and add the first part to the end?

```python
def splitArr(arr, n, k):
    for i in range(0, k):
        x = arr[0]
        for j in range(0, n-1):
            arr[j] = arr[j + 1]
         
        arr[n-1] = x
         
arr = [12, 10, 5, 6, 52, 36]
n = len(arr)
position = 2
 
splitArr(arr, n, position)
 
for i in range(0, n):
    print(arr[i], end = ' ')
```

5. Write a Python Program to check if given array is Monotonic?

```python
def isMonotonic(A):
    x, y = [], []
    x.extend(A)
    y.extend(A)
    x.sort()
    y.sort(reverse=True)
    if(x == A or y == A):
        return True
    return False
 
A = [6, 5, 4, 4]

print(isMonotonic(A))
```
