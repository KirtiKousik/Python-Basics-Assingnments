1. Write a Python program to check if the given number is a Disarium Number?

```python
def is_disarium(num):
    sum = 0
    for i in range(len(str(num))):
        sum += int(str(num)[i]) ** (i + 1)
    return sum == num
```

```python
is_disarium(135)
```

```python
is_disarium(175)
```

```python
is_disarium(145)
```

2. Write a Python program to print all disarium numbers between 1 to 100?

```python
def in_range_disarium(start_range, end_range):
    for num in range(start_range,end_range+1):
        sum = 0
        for i in range(len(str(num))):
            sum += int(str(num)[i]) ** (i + 1)
        if sum == num:
            print(sum,end=' ')
```

```python
in_range_disarium(1,100)
```

3. Write a Python program to check if the given number is Happy Number?

```python
def is_Happy_num(n):
  past = set()
  while n != 1:
        n = sum(int(i)**2 for i in str(n))
        if n in past:
            return False
        past.add(n)
  return True
print(is_Happy_num(7))
print(is_Happy_num(932))
print(is_Happy_num(6))
print(is_Happy_num(49))
```

4. Write a Python program to print all happy numbers between 1 and 100?

```python
def is_Happy_num(n):
  past = set()
  while n != 1:
        n = sum(int(i)**2 for i in str(n))
        if n in past:
            return False
        past.add(n)
  return True
print('The happy Numbers between 1 and 100 are:')
for i in range(1,101):
    if is_Happy_num(i) == True:
        print(i,end=' ')
```

5. Write a Python program to determine whether the given number is a Harshad Number?

```python
def is_harshad_number(n):
    sum = 0
    for i in str(n):
        sum += int(i)
    if n%sum == 0:
        return 'Harshad Number'
    else:
        return "Not a Harshad Number"
```

```python
n = int(input('Enter a number: '))
print(f'The number {n} is {is_harshad_number(n)}')
```

6. Write a Python program to print all pronic numbers between 1 and 100?

```python
# to check in range if a number is pronic number or not
for n in range(1,100):
    check = False
    for i in range(n):
        if (i * (i+1)) == n:
            check = True
            break
    if check:
        print(n,end=' ')
```

```python
# To check a number is Pronic Number or not
def is_pronic_number(n):
    check = False
    for i in range(n):
        if (i * (i+1)) == n:
            check = True
            break

    if check == True:
        print(f'{n} is a pronic number')
    else:
        print(f'{n} is not a pronic number')

is_pronic_number(420)
```
