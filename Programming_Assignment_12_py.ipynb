1. Write a Python program to Extract Unique values dictionary values?

```python
d = {'k': [1,2,3], 'i': [2,2,2], 'r': 'biswal'}

print('The original dictionary is:\n',d,'\n')
op = []
for i in d.items():
    op.append(i[1])

print('The unique value list is:\n',op)
```

2. Write a Python program to find the sum of all items in a dictionary?

```python
d = {'a': 100, 'b': 200, 'c': 300}
print('The original dictionary is:\n',d,'\n')
sum = 0
for i in d.values():
    sum += i

print(f'The sum of all items in the dictionary is {sum}')
```

4. Write a Python program to convert key-values list to flat dictionary?

```python
org_dict = {'month' : [1, 2, 3],'name' : ['Jan', 'Feb', 'March']}
 
print("The original dictionary is :", org_dict,'\n')

x=list(org_dict.values())
a=x[0]
b=x[1]
new_dict = dict()
for i in range(0,len(a)):
    new_dict[a[i]]=b[i]

print("Flattened dictionary :",new_dict)
```

5. Write a Python program to insertion at the beginning in OrderedDict?

```python
from collections import OrderedDict
dic1 = OrderedDict([('A', '100'), ('B', '200'), ('C', '300')])
insrt = OrderedDict([("D", '400')])
  
op = OrderedDict(list(insrt.items()) + list(dic1.items()))

print ("Resultant Dictionary :")
print(op)
```

6. Write a Python program to check order of character in string using OrderedDict()?

```python
from collections import OrderedDict 
def checkOrder(string, pattern): 
    dic = OrderedDict.fromkeys(string) 
    ptr = 0
    for key,value in dic.items(): 
        if (key == pattern[ptr]): 
            ptr = ptr + 1
        if (ptr == (len(pattern))): 
            return 'True'
    return 'False'
```

```python
string = 'Study tonight'
pattern = 'stu'
print (checkOrder(string,pattern))
```

```python
string2= 'Welcome'
pattern2= 'cm'
print (checkOrder(string2,pattern2))
```

7. Write a Python program to sort Python Dictionaries by Key or Value?

```python
# sorted by keys
from collections import OrderedDict
 
dict = {'ravi': '10', 'rajnish': '9',
        'sanjeev': '15', 'yash': '2', 'suraj': '32'}
dict1 = OrderedDict(sorted(dict.items()))
print(dict1)
```

```python
# sorting by value
from collections import OrderedDict
import numpy as np
 
dict = {'ravi': 10, 'rajnish': 9,
        'sanjeev': 15, 'yash': 2, 'suraj': 32}
print('Original Dictionary:\n',dict,'\n')
 
keys = list(dict.keys())
values = list(dict.values())
sorted_value_index = np.argsort(values)
sorted_dict = {keys[i]: values[i] for i in sorted_value_index}
 
print('Sorted dictionary:\n',sorted_dict)
```
