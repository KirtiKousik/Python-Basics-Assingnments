1. Write a Python program to find words which are greater than given length k?

```python
k = int(input('Enter k: '))
sentence = input('Enter a string: ').split()
print(f'Words with the length more than {k}(k) are:')
for i in sentence:
    if len(i) > k:
        print(i,end='\t')
```

2. Write a Python program for removing i-th character from a string?

```python
def remove_ith(s,n):
    first = s[:n]
    last = s[n+1:]
    return first+last

s = input("Enter a string: ")
n = int(input('Enter the index you want to remove: '))
print('New string:',remove_ith(s,n))
```

3. Write a Python program to split and join a string?

```python
s = 'I Love Python'
print('The original string is:\n',s)

#spliing the string
print('The split string is: ')
l = s.split()
print(l)

#joining the string
print('The joined string is: ')
j = '-'.join(l)
print(j)
```

4. Write a Python to check if a given string is binary string or not?

```python
def is_binary(s):
    check = False
    for i in s:
        if i == '1' or i == '0':
            check = True
        else:
            check = False
            break

    if check:
        print('Binary String')
    else:
        print('Not Binary String')

is_binary('101011110011')
is_binary('kirti1011')
is_binary("101Pyt101hon101")
```

5. Write a Python program to find uncommon words from two Strings?

```python
def uncommon_words(s1,s2):

    op = []
    for i in s1.split():
        if i not in s2.split():
            op.append(i)
    for i in s2.split():
        if i not in s1.split():
            op.append(i)

    return list(set(op))

s1 = 'Kirti Loves Python very much'
s2 = 'Ribu Loves Python too much'
uncommon_words(s1,s2)
```

6. Write a Python to find all duplicate characters in string?

```python
s = input('Enter a string: ')

op= []
for i in s:
    if s.count(i) > 1:
        op.append(i)
if op != []:
    print('The duplicate characters are:\n',list(set(op)))
else: 
    print('There are no duplicate characters is this string.')
```

7. Write a Python Program to check if a string contains any special character?

```python
n = 'KirtiBiswal@'
n.split()
c=0
s='[@_!#$%^&*()<>?/\|}{~:]'
for i in range(len(n)):
    if n[i] in s:
      c+=1

if c:
    print("string contains special character.")
else:
    print("string does not contain special character.")
```
