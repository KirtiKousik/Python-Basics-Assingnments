1. Write a Python program to find sum of elements in list?

```python
l = [1,2,3,4,5,6]
sum = 0
for i in l:
    sum += i
print('list: ',l)
print('sum of list elements: ',sum) 
```

2. Write a Python program to Multiply all numbers in the list?

```python
l = [1,2,3,4,5,6]
product = 1
for i in l:
    product *= i
print('list: ',l)
print('Product of list elements: ',product)
```

3. Write a Python program to find smallest number in a list?

```python
l = [12,2,3,4,5,6,11]
min = l[0]
for i in l:
    if i <= min:
        min = i

print("Smallest element is:",min)
```

```python
# or 
l = [12,2,3,4,5,6,11]
print("Smallest element is:", sorted(l)[0])
```

4. Write a Python program to find largest number in a list?

```python
l = [1,32,3,4,5,6,11,3]
max = 0
for i in l:
    if i > max:
        max = i

print('The largest list element is',max)
```

```python
# or
l = [1,32,3,4,5,6,11,3]
print("The largest Number is:", sorted(l)[-1])
```

5. Write a Python program to find second largest number in a list?

7. Write a Python program to print even numbers in a list?

```python
l = [1,2,3,4,5,6,7,8,9]
print('The even numbers in the list are:')
for i in l:
    if i%2==0:
        print(i,end=' ')
```

8. Write a Python program to print odd numbers in a List?

9. Write a Python program to Remove empty List from List?

```python
# example 1
l = [1,[],3,[],[],5,[2,3,4],6,[]]
new_list = list()
for i in l:
    if i != []:
        new_list.append(i)
print('The empty list removed list: ',new_list)
```

```python
# example 2
l2 = [5, 6, [], 3, [], [], 9]
new_list = list()
for i in l2:
    if i != []:
        new_list.append(i)
print('The empty list removed list: ',new_list)
```

10. Write a Python program to Cloning or Copying a list?

```python
# or
def list_cloning(l):
    clone_list = list()
    for i in l:
        clone_list.append(i)
    return clone_list

original_list = [1,2,3,4,5,6,7,8,9]
print("Cloned List:", list_cloning(original_list))
```

```python
# or 
original_list = [1,2,3,4,5,6,7,8,9]
clone_list = original_list

print(f'original_list: {original_list}')
print(f'clone_list: {clone_list}')
```
